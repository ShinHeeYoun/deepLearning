```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Input, Concatenate
from tensorflow.keras.models import Model

main_input = Input(shape=(1024,))
aux_input = Input(shape=(10,))

D1 = Dense(128, activation='relu') (main_input)
D2 = Dense(64, activation='relu') (D1)

M1 = Concatenate() ([aux_input, D2])
main_output = Dense(1, activation='softmax') (M1)
aux_output = Dense(10, activation='softmax') (D2)

#   ( )
model = Model(inputs=[main_input, aux_input], outputs=[main_output, aux_output])
model.summary()

# 
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model.jpg', show_shapes=True)
```
