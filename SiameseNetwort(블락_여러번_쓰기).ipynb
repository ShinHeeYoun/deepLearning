```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Input, Concatenate, Conv2D, MaxPooling2D, Flatten, Subtract
from tensorflow.keras.models import Model

input1 = Input(shape=(224, 224, 3))
input2 = Input(shape=(224, 224, 3))

# 레이어 같이 쓰기
f_conv = Conv2D(16, (3,3), padding='same')
conv1 = f_conv (input1)
conv2 = f_conv (input2)
#conv1 = Conv2D(16, (3,3), padding='same') (input1)
#conv2 = Conv2D(16, (3,3), padding='same') (input2)

output = Subtract() ([conv1, conv2])

# 모델 구성 (입력값들과 출력값들)
model = Model(inputs=[input1, input2], outputs=[output])
model.summary()

# 출력
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model.jpg', show_shapes=True)
```
