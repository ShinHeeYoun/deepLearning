```python
from tensorflow.keras.layers import Input, Concatenate
from tensorflow.keras.models import Model

#    3
input1 = Input(shape=(28*28,))
input2 = Input(shape=(28*28,))
input3 = Input(shape=(28*28,))

# 
# input1 + input2 --> output1
# output1 + input3 --> output2
output1 = Concatenate() ([input1, input2])
output2 = Concatenate() ([output1, input3])

#   ( )
model = Model(inputs=[input1, input2, input3], outputs=[output2])
model.summary()

#  
from tensorflow.keras.utils import plot_model
plot_model( model, to_file='model2.jpg', show_shapes=True)
```
